```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

import keras
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
np.random.seed(0)
n_pts = 500
X, y = datasets.make_circles(n_samples = n_pts, random_state = 123, noise = 0.1, factor = 0.2)
print(X) 
print(y)

# Selectively plotting the outer region i.e. label 0 and performing a Boolean check
plt.scatter(X[y == 0, 0], X[y == 0, 1])
# Selectively plotting the inner region i.e. label 1 and performing a Boolean check
plt.scatter(X[y == 1, 0], X[y == 1, 1])
```

```python
# the parameter values used are based on their performance in the Tensorflow Playground
model = Sequential()
# hidden layer = 4 nodes & input layer = 2 nodes
model.add(Dense(4, input_shape = (2,), activation = 'sigmoid'))
# output layer = 1 node
model.add(Dense(1, activation = 'sigmoid'))
# setting the loss function to binary crossentropy since we're distinguishing between two different classes
model.compile(Adam(lr = 0.01), 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# the history object returned and stored in 'h' keeps track of the training loss and the accuracy for each epoch
h = model.fit(x = X, y = y, verbose = 1, batch_size = 20, epochs = 100, shuffle = "true")
```

```python
plt.plot(h.history['accuracy'])
plt.legend(['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
 
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    grid = np.c_[xx.ravel(), yy.ravel()]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
plot_decision_boundary(X, y, model)
plt.scatter(X[y == 0, 0], X[y == 0, 1])
plt.scatter(X[y == 1, 0], X[y == 1, 1])
plot_decision_boundary(X, y, model)
plt.scatter(X[y == 0, 0], X[y == 0, 1])
plt.scatter(X[y == 1, 0], X[y == 1, 1])
 
x = 0
y = 0.75
 
point = np.array([[x, y]])
predict = model.predict(point)
plt.plot([x], [y], marker = 'o', markersize = 10, color = "red")
print("Prediction is: ", predict)
```
